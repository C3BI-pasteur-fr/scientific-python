# Solution â€” k-means (k=2) + Rand Index

1) Load data
2) Standardize and fit k-means
3) Rand Index vs `group`.


```python
# 1. Load dataset
import pandas as pd

df = pd.read_csv("../data/day5_example.csv")
y = df['group']
X = df[['Gene1','Gene2','Gene3']]
df.head()
```

```python
# 2. Fit k-means (k=2) after standardizing
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

scaler = StandardScaler()
Xz = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=2, random_state=0, n_init=10)
labels = kmeans.fit_predict(Xz)
labels
```

```python
# 3. Rand Index vs true `group`
from sklearn.metrics import rand_score

ri = rand_score(y, labels)
ri
```
