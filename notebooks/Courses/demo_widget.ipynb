````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
def chapeau(x,y):
    r = np.sqrt(x**2 + y**2)
    z = np.sin(r)
    return z
````

````python

func, x_min, x_max, y_min, y_max, n, = chapeau, -5, 5, -5, 5, 50
plot='surface'
cmap='plasma'

VX = np.linspace(x_min, x_max, n)
VY = np.linspace(y_min, y_max, n)
    
X, Y = np.meshgrid(VX, VY) 
    
Z = func(X, Y)
    
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.view_init(40, -70)
plot = ax.plot_surface(X,Y,Z, cmap=cmap)
fig.colorbar(plot)
    
````
